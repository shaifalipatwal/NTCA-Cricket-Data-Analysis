### **Feature Engineering**

```python
# Batting Consistency Metrics:This metric can help quantify how consistent a player is across different matches.
batting_df['Consistency_Score'] = batting_df['25\'s']+ batting_df['50\'s'] + batting_df['75\'s'] + batting_df['100\'s'] - batting_df['0']
```

```python
# Boundary Percentage: This metric provides insights into the player's playing style.
batting_df["Boundary%"]= (batting_df['4\'s'] + batting_df['6\'s']) / batting_df['Runs'] * 100
```

```python
# Runs per Match: This metric can indicate a player's performance consistency over all matches played.
batting_df["Runs_per_match"]= batting_df["Runs"] / batting_df["Mat"]
```

```python
# Runs per Innings: This feature is useful for comparing players based on how many runs they score on average when they get a chance to bat.
batting_df['Runs_per_Innings'] = batting_df['Runs'] / batting_df['Ins']
```

```python
# No. of Not-outs to Matches Ratio: It indicates the player's ability to finish the game.
batting_df['Not_out_Ratio'] = batting_df['No'] / batting_df['Mat']
```

```python
batting_df["Avg"].fillna(0,inplace=True)
batting_df["Boundary%"].fillna(0,inplace=True)
batting_df["Runs_per_Innings"].fillna(0,inplace=True)
```

### **Machine Learning Models**

Applying Random Forest model to predict the runs scored by players

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X= batting_df.drop(["Runs",'Player', 'Group', 'Team'],axis='columns')
y= batting_df["Runs"]
```

```python
X.head(1)
```

```python
y.head(1)
```

```python
# Splitting the dataset into training and test sets
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
model= RandomForestRegressor(n_estimators=100, random_state=10)
model.fit(X_train,y_train)
```

```python
# Making predictions
y_pred = model.predict(X_test)
```

```python
# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse
```

```python
r2
```

```python
model.score(X_test,y_test)   # r-squared value to check the accuracy of our model
```

```python
# Create a DataFrame with actual and predicted values ("Runs")
comparison_df = pd.DataFrame({'Player': X_test.index.map(batting_df['Player']),'Actual Runs': y_test, 'Predicted Runs': y_pred})

# Display the first few rows of the comparison
comparison_df.head(10)
```
